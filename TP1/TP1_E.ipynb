### La version python de `dicho`qui permettra de trouver les solutions de l'équation $f(x)=0$ avec $eps=0.001$.

```python
def dicho(a,b,f,eps):
    '''version tant que '''
    
    while (b-a)>eps:
        #Moyenne
        M=(a+b)/2
        if f(M)==0:
            return M
        if f(a)*f(M)<0 :
            b=M
        else :
            a=M
    return M
```

```python
f=lambda x: x**3+x**2-3*x-3
dicho(-2,-1.5,f,0.001)
```

```python
f=lambda x: x**3+x**2-3*x-3
dicho(-1.5,-0.5,f,0.001)
```

### La version tant que qui renvoi la solution avec le nombre d'itération


```python
## Ecrire une fonction qui permet de renvoyer le nombre d'itération nécessaire à la méthode de dichotomie pour calculer
## la solution à 10^-p près 
def dichow(a,b,f,eps):
    """version tant que qui renvoila solution avec le nombre d'''itération """
    m=(a+b)/2
    err=abs(b-a)
    n=0
    while err>eps:
        n=n+1
        if f(m)==0 :
            break
        elif f(a)*f(m)<0 :
            b=m
        else :
            a=m
        m=(a+b)/2
        err=abs(b-a)
    return m,n
```

- $\alpha_2 \in ]-1.5;-0.5[$ 
- $\alpha_3 \in ]1.5;2[$

>1) On considère la fonction $f(x)=e^{-x}-x$. Ecrire puis appliquer  une fonction Matlab que l'on appellera `newton.m` à la >fonction $f$, en prenant $x0=0$, $eps=0.001$ .
>
>2) Montrer que l'équation $f(x)=0$ admet une solution unique que l'on peut obtenir en utilisant la méthode de Newton à partir >de $x0=1$.

#### 1) <h4>

### Solution Python

```python

def newton(f,fprime,a,eps):
    x1=a
    x2=x1-(f(x1)/fprime(x1))
    while abs(x1-x2)>eps :
        x1=x2
        x2=x1-(f(x1)/fprime(x1))
        m=x2
    return m
```

```python
import math
from sympy import *
from sympy.abc import x

f=lambda x: math.exp(-x)-x
fprime=lambda x: -math.exp(-x)-1
newton(f,fprime,0,0.001)
```

```python
newton(f,fprime,1,0.001)
```

### Interprétations

La méthode de newton est une cas particulier de la méthode de point fixe.


Avantages de la methode de newton :
* L'avantage majeur de la méthode de Newton par rapport à une méthode de point fixe par exemple est sa vitesse de convergence d'ordre 2.

Inconvénients de la methode de newton :
* dépend de la fonction f et son dérivé.
